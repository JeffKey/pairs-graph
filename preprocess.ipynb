```python
import yfinance as yf
import pandas as pd
import numpy as np

# Fetch list of smartbeta ETFs. This is a placeholder; replace with an actual source or list.
etf_list = ["ETF1", "ETF2", ...]  # Placeholder list

# Get trading volume for each ETF and pick the top 25 by volume
volumes = {}
for etf in etf_list:
    data = yf.download(etf, period="5d", interval="2m")
    volumes[etf] = data['Volume'].sum()

top_25_etfs = sorted(volumes, key=volumes.get, reverse=True)[:25]

print(top_25_etfs)
```

```python

top_25_etfs = ['USMV', 'SPMV', 'EFAV', 'SPLV', 'XMLV', 'VUG', 'MTUM', 'VFMO', 
        'PDP', 'VFMF', 'VTV', 'SCHV', 'RPV', 'IUSV', 'IVOV', 'QUAL', 
        'SPHQ', 'JHML', 'DGRW', 'QUS', 'SMLF', 'VIOO', 'IJS', 'IJR', 'VB']
# Collect 2-minute historical data for these ETFs
historical_data = {etf: yf.download(etf, period="60d", interval="2m") for etf in top_25_etfs}

# Compute relative price for pairs trading
relative_prices = pd.DataFrame(index=historical_data[top_25_etfs[0]].index)
for i, etf1 in enumerate(top_25_etfs):
    for j, etf2 in enumerate(top_25_etfs):
        if i < j:
            pair_name = f"{etf1}/{etf2}"
            relative_prices[pair_name] = historical_data[etf1]['Close'] / historical_data[etf2]['Close']
```

Take all of the pertinent information and structure it into a graph similar to 

```python

# Structure the data in graph format
graph_data = {
    'nodes': top_25_etfs,
    'edges': [(etf1, etf2) for i, etf1 in enumerate(top_25_etfs) for j, etf2 in enumerate(top_25_etfs) if i < j],
    'edge_features': relative_prices.to_dict(orient='list'),
    'node_features': {etf: historical_data[etf]['Close'].tolist() for etf in top_25_etfs}
}

print(graph_data)
```
